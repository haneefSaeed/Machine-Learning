MATPLOTLIB

```python
#importing

import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
print(x, y , z)
```

```python
## plotting the figure

plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, z)
plt.show()
```

```python
plt.plot(x, y)
plt.xlabel("angle")
plt.ylabel("sine value")
plt.title("Sine wave")
plt.show()
```

Parabola

```python
x = np.linspace(-10, 10, 20)
y = x**2
plt.plot(x, y, 'g.')
```

```python
x = np.linspace(-5, 5, 500)
plt.plot(x, np.sin(x), 'r-')
plt.plot(x, np.cos(x) , 'g--')
```

FIGURES


```python
fig  = plt.figure()
ax = fig.add_axes([100,100,1,1])
lang = ['EN', "AR", "FR", 'GR']
people = [100, 20, 40, 80]
ax.bar(lang, people)

fig.show()
```

```python
fig  = plt.figure()
ax = fig.add_axes([100,100,1,1])
lang = ['EN', "AR", "FR", 'GR']
people = [100, 20, 40, 80]
ax.pie( people, labels=lang, autopct='%1.1f%%')
```
