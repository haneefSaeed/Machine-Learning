Data Visualizing using Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
## total bill vs tip dataset
## how much total tip will give for total bill at
tips  = sns.load_dataset('tips')
tips.head()
```

```python
## default theme for plotting
sns.set_theme()
```

```python
# visualize
sns.relplot(data=tips, x ='total_bill', y='tip', col='time', hue='smoker', style='smoker', size='size')
```

```python
## iris dataset

iris = sns.load_dataset('iris')

iris.head()
```

```python
sns.scatterplot(data=iris, x='sepal_length', y='petal_width', hue='species' )
```

```python
## titanic dataset
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.shape
```

```python
## count plot
sns.countplot(x='sex', data=titanic)
```

```python
## Count plot
sns.countplot(x='class', data=titanic, hue='class')
```

```python
## bar chart
sns.barplot(x='sex', y='survived', hue='class', data=titanic)
```

```python
## house price dataset

from sklearn.datasets import fetch_california_housing
```

```python
load_house  = fetch_california_housing()

house = pd.DataFrame(load_house.data, columns=load_house.feature_names)
house['Price'] = load_house.target*1000
house.head()
```

Distribution Plot and Matrix

```python
sns.distplot(house['Price'])
```

Correlation Matrix
+ ve correlation
- ve correlation

```python
# Heat Map

correlation = house.corr()


#construction of heat Map

plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size': 8}, cmap='Blues')
```
