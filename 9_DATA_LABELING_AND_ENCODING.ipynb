Converting Label into Numeric

```python
import numpy as np
import pandas as pd
import sklearn.datasets
import seaborn as sns

from sklearn.preprocessing import LabelEncoder
```

```python
cancerDataset = sklearn.datasets.load_breast_cancer()
irisDataset = sklearn.datasets.load_iris()

cdf = pd.DataFrame(cancerDataset.data, columns =cancerDataset.feature_names)
cdf.head()
```

```python

idf = pd.DataFrame(irisDataset.data, columns = irisDataset.feature_names)
idf.head()
```

```python
cdf.shape
# cdf['diagnosis'] = ['M' if cdf['worst texture'][x] >18 else "N" for x in range(569)]

## same as
cdf['diagnosis'] = np.where(cdf['worst texture'] > 22 , "M", "N")
cdf.head()
```

```python
cdf['diagnosis'].value_counts()
```

```python
# Load Label Encoder

label_encoder = LabelEncoder()
```

```python
labels = label_encoder.fit_transform(cdf['diagnosis'])
print(labels)

cdf['diagnosis_numeric'] = labels
```

```python
cdf.head()
```

```python
# iris data label encoding

idf.head()
idf['sepal length (cm)'].value_counts()

# creating a classification column imaginary

idf['color'] = ['Green' if idf['sepal length (cm)'][x]>=4 and idf['sepal length (cm)'][x]<=5 else "Blue " if idf['sepal length (cm)'][x]>=5 and idf['sepal length (cm)'][x]<6 else "Black"  for x in range(150) ]

idf.head()

idf['color'].value_counts()
```

```python
eyeColorLabels = label_encoder.fit_transform(idf['color'])
print(eyeColorLabels)
```

```python
# assign the new labels
idf['col_type'] = eyeColorLabels
idf.head()
```
