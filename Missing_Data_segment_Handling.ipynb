Methods of Handling Missing Data

* Dropping


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('Placement_Dataset.csv')
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

In case, the dataset is small, we can impute the missing data by using centeral tendencies.
1. Mean
2. Mode
3. Median

We cannot use random, we just have to observe first which one fits best.


```python
# check the distribution.
sns.distplot(dataset.salary)
```

Looking at distribution, we see that we have 2 outliers around 0.7 and 0.9. that means if we use the mean, we will have the value that is not accurate enough, because the distribution is skewed.


in case the distribution is normal, we can use mean.
in the skewed case we can use median or mode.

```python
dataset.head()
```

```python
#replace using median /mode value

# dataset['salary'].fillna(dataset['salary'].median(), inplace=True)
dataset['salary'].fillna(dataset['salary'].mode(), inplace=True)
# dataset['salary'].fillna(median, inplace=True)
```
